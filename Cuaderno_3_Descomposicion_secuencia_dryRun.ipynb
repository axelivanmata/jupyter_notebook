## 1. Descomposición del problema
Dividir un problema grande en partes pequeñas y manejables.

### Ejemplo: 
**Descomposición:** Calcular tiempo en segundos
1. Leer horas.
2. Leer minutos.
3. Leer segundos.
4. Convertir horas → segundos (horas × 3600).
5. Convertir minutos → segundos (minutos × 60).
6. Sumar todos los segundos.
7. Mostrar resultado.

```python
horas = int(input("Horas: "))
minutos = int(input("Minutos: "))
segundos = int(input("Segundos: "))
total = horas * 3600 + minutos * 60 + segundos
print("Total de segundos:", total)
```

## 5. Ejemplo dry-run

> Calcular el índice de masa corporal (IMC) a partir del peso y la estatura.

```python
peso = float(input('Peso en kg: '))
estatura = float(input('Estatura en metros: '))
imc = peso / (estatura ** 2)
print('IMC =', imc)
```
