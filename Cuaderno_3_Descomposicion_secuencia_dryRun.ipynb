## Condicionales: simples y compuestas

### Conceptos clave
- `if` (condicional simple)
- `if`/`else` (condicional compuesta)
- `if`/`elif`/`else` (condicional mĂşltiple)
- Operadores relacionales: `>`, `<`, `>=`, `<=`, `==`, `!=`
- Operadores lĂłgicos: `and`, `or`, `not`


```python
# Ejemplo: condicional simple
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
```

```python
# Ejemplo: condicional compuesta
nota = 65
if nota >= 60:
    print("Aprobado")
else:
    print("Reprobado")
```

```python
# Ejemplo: condicional mĂşltiple
temperatura = 30
if temperatura > 35:
    print("Hace mucho calor")
elif temperatura >= 25:
    print("Clima agradable")
else:
    print("Hace frĂ­o")
```

### Ejemplo: Signo de un nĂşmero
**Instrucciones:** Pide al usuario un nĂşmero e imprime si es *positivo*, *negativo* o *cero*.

> Sugerencia: convierte la entrada a `float` y usa `if`/`elif`/`else`.

## Bucles: por contador (`for`) y por condiciĂłn (`while`)

### `for` (con contador)
- Usamos `range(inicio, fin, paso)`; "fin" no se incluye.
- Ideal cuando sabemos cuĂĄntas iteraciones necesitamos.


```python
# Ejemplo con for
for i in range(5):
    print("IteraciĂłn:", i)
```

### `while` (con condiciĂłn)
- Se ejecuta **mientras** la condiciĂłn sea verdadera.
- Ătil cuando no sabemos cuĂĄntas repeticiones habrĂĄ.


```python
# Ejemplo con while
contador = 1
while contador <= 5:
    print("Contando:", contador)
    contador += 1
```

## Control de flujo en bucles: `break`, `continue` y `pass`
- `break`: sale del ciclo inmediatamente.
- `continue`: salta a la siguiente iteraciĂłn.
- `pass`: marcador de posiciĂłn; no hace nada.


```python
# break: detiene el bucle al llegar a 5
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue: imprime solo impares
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass: marcador de lugar (Ăştil al diseĂąar la estructura)
for i in range(3):
    pass  # AquĂ­ podrĂ­as implementar algo luego
print("Se ejecutĂł el bucle con pass")
```

## IntroducciĂłn a listas simples
Una **lista** es una colecciĂłn ordenada y modificable. Se define con corchetes `[]`.

### Operaciones bĂĄsicas
- Crear una lista
- Acceder por Ă­ndice
- Modificar elementos
- Agregar/eliminar
- Recorrer con `for`


```python
# Crear y manipular listas
frutas = ["manzana", "pera", "naranja"]
print("Lista inicial:", frutas)

# Acceso por Ă­ndice
print("Primera fruta:", frutas[0])

# Modificar elemento
frutas[1] = "banano"
print("DespuĂŠs de modificar:", frutas)

# Agregar y eliminar
frutas.append("melĂłn")
frutas.remove("naranja")
print("DespuĂŠs de append y remove:", frutas)

# Recorrer la lista
for fruta in frutas:
    print("Fruta:", fruta)
```
