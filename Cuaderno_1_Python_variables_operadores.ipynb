## 1. Introducción: ¿Qué es un dato en Python?
- Python es **tipado dinámico**: no declaras el tipo de la variable; el intérprete lo infiere en tiempo de ejecución.
- Todo en Python es un **objeto**.

**Actividad rápida:** ejecuta la celda siguiente y observa los tipos con `type()`.


```python
x = 10            # int
nombre = 'Ana'   # str
pi = 3.1416      # float
activo = True    # bool

type(x), type(nombre), type(pi), type(activo)
```

## 2. Variables
Una **variable** es un nombre que referencia un valor en memoria.

**Reglas de nombres:**
- Letras, dígitos y guion bajo; no comenzar con dígito.
- Sensibles a mayúsculas/minúsculas.
- Evitar palabras reservadas (`if`, `for`, `class`, etc.).

**Asignación y reasignación:**


```python
edad = 25
edad = edad + 1
edad
```

**Asignación múltiple y desempaquetado:**


```python
a, b, c = 10, 20, 30
x = y = 5
a, b, c, x, y
```

## 3. Constantes (convención)
Python no tiene *constantes verdaderas*. Por convención, se escriben en MAYÚSCULAS.
Opcionalmente se puede usar `Final` para indicar intención al tipo estático.


```python
from typing import Final
PI: Final = 3.1416
MAX_USUARIOS: Final = 100
PI, MAX_USUARIOS
```

## 4. Tipos de datos básicos y *casting*
- `int`, `float`, `str`, `bool`
- Conversión entre tipos con `int()`, `float()`, `str()`


```python
int('10'), float('3.14'), str(123), bool(0), bool(1)
```

## 5. Operadores aritméticos
`+`, `-`, `*`, `/` (float), `//` (entera), `%` (módulo), `**` (potencia).


```python
(10 + 5) * 2, 7/2, 7//2, 7%3, 2**5
```

## 6. Operadores relacionales y lógicos
Relacionales: `==`, `!=`, `>`, `<`, `>=`, `<=`

Lógicos: `and`, `or`, `not`


```python
nota = 75
aprobado = nota >= 60
edad = 19
tiene_dpi = True
aprobado, (edad >= 18 and tiene_dpi), (not False)
```

## 7. Operadores de asignación compuesta
Permiten actualizar una variable con una operación en el mismo paso.


```python
x = 5
x += 2   # 7
x *= 3   # 21
x
```

## 9. Ejemplo
**Enunciado:**
1) Pide al usuario el radio de un círculo.
2) Calcula **área** y **circunferencia** usando `PI`.
3) Imprime si el área es "grande" (> 50) y si la circunferencia es "larga" (> 30).


PI = 3.1416
try:
    radio = float(input('Ingrese el radio: '))
    area = PI * (radio ** 2)
    circ = 2 * PI * radio
    print('Área =', area)
    print('Circunferencia =', circ)
    if area > 50:
        print('Área grande')
    else:
        print('Área pequeña')
    if circ > 30:
        print('Circunferencia larga')
    else:
        print('Circunferencia corta')
except ValueError:
    print('Entrada inválida: ingrese un número.')
