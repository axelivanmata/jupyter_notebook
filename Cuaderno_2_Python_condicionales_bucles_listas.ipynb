## Condicionales: simples y compuestas

### Conceptos clave
- `if` (condicional simple)
- `if`/`else` (condicional compuesta)
- `if`/`elif`/`else` (condicional múltiple)
- Operadores relacionales: `>`, `<`, `>=`, `<=`, `==`, `!=`
- Operadores lógicos: `and`, `or`, `not`


```python
# Ejemplo: condicional simple
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
```

```python
# Ejemplo: condicional compuesta
nota = 65
if nota >= 60:
    print("Aprobado")
else:
    print("Reprobado")
```

```python
# Ejemplo: condicional múltiple
temperatura = 30
if temperatura > 35:
    print("Hace mucho calor")
elif temperatura >= 25:
    print("Clima agradable")
else:
    print("Hace frío")
```

### Ejemplo: Signo de un número
**Instrucciones:** Pide al usuario un número e imprime si es *positivo*, *negativo* o *cero*.

> Sugerencia: convierte la entrada a `float` y usa `if`/`elif`/`else`.


```python
numero = float(input("Ingresa un número: "))
if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:
    print("Cero")
```

## Bucles: por contador (`for`) y por condición (`while`)

### `for` (con contador)
- Usamos `range(inicio, fin, paso)`; "fin" no se incluye.
- Ideal cuando sabemos cuántas iteraciones necesitamos.


```python
# Ejemplo con for
for i in range(5):
    print("Iteración:", i)
```

### `while` (con condición)
- Se ejecuta **mientras** la condición sea verdadera.
- Útil cuando no sabemos cuántas repeticiones habrá.


```python
# Ejemplo con while
contador = 1
while contador <= 5:
    print("Contando:", contador)
    contador += 1
```

## Control de flujo en bucles: `break`, `continue` y `pass`
- `break`: sale del ciclo inmediatamente.
- `continue`: salta a la siguiente iteración.
- `pass`: marcador de posición; no hace nada.


```python
# break: detiene el bucle al llegar a 5
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue: imprime solo impares
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass: marcador de lugar (útil al diseñar la estructura)
for i in range(3):
    pass  # Aquí podrías implementar algo luego
print("Se ejecutó el bucle con pass")
```

## 4️⃣ Introducción a listas simples
Una **lista** es una colección ordenada y modificable. Se define con corchetes `[]`.

### Operaciones básicas
- Crear una lista
- Acceder por índice
- Modificar elementos
- Agregar/eliminar
- Recorrer con `for`


```python
# Crear y manipular listas
frutas = ["manzana", "pera", "naranja"]
print("Lista inicial:", frutas)

# Acceso por índice
print("Primera fruta:", frutas[0])

# Modificar elemento
frutas[1] = "banano"
print("Después de modificar:", frutas)

# Agregar y eliminar
frutas.append("melón")
frutas.remove("naranja")
print("Después de append y remove:", frutas)

# Recorrer la lista
for fruta in frutas:
    print("Fruta:", fruta)
```
