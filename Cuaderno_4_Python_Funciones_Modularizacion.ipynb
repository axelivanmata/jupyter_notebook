## Fundamentos de las funciones

### ¿Qué es una función?
Una **función** es un bloque de código reutilizable que realiza una tarea específica.

### Sintaxis básica
```python

    """Docstring: describe qué hace la función.
    Parámetros:
      par1: descripción
      par2: descripción
    Retorna: tipo y significado
    """
    # cuerpo
    return resultado
```


````python
# Ejemplo simple de función
def saludar():
    """Imprime un saludo simple.
    No recibe parámetros y no retorna valor (retorna None implícitamente).
    """
    print("Hola desde una función")

saludar()
````

## Parámetros y valores de retorno

### Parámetros (entrada)
Los **parámetros** permiten pasar datos a la función.
- Posicionales
- Con nombre (keywords)
- Con valores por defecto

### Retorno (salida)
`return` envía un valor (o varios) de vuelta al llamador.


````python
# Ejemplos de parámetros y retorno
def saludar_a(nombre, saludo="Hola"):
    return f"{saludo}, {nombre}!"

print(saludar_a("Axel"))
print(saludar_a(nombre="Clase", saludo="Bienvenidos"))

# Retorno de múltiples valores (tuplas)
def operaciones_basicas(a, b):
    resta = a / b
    return suma, resta

s, r = operaciones_basicas(10, 3)
print("Suma:", s, "— Resta:", r)
````

## Modularización
Dividir un problema en **módulos** (funciones) más pequeños mejora la legibilidad, mantenimiento y pruebas.

### Ejemplo: Antes vs. Después
**Sin modularización**:
```python
a = int(input("A: "))
b = int(input("B: "))
c = int(input("C: "))
prom = (a + b + c) / 3
print("Promedio:", prom)
```
**Con modularización**:
```python
def promedio3(a, b, c):
    return (a + b + c) / 3

def leer_entero(msg):
    return int(input(msg))

a = leer_entero("A: ")
b = leer_entero("B: ")
c = leer_entero("C: ")
print("Promedio:", promedio3(a, b, c))
```


## Abstracción
La **abstracción** oculta los detalles internos y expone una interfaz simple.
Piensa en "qué hace" la función más que en "cómo lo hace".


````python
# Ejemplo de abstracción: normalizar texto
import re

def normalizar_texto(s):
    """
    Retorna la cadena en minúsculas, sin múltiplos espacios y sin espacios a los extremos.
    """
    s = s.lower().strip()
    s = re.sub(r"\s+", " ", s)
    return s

print(normalizar_texto("   Hola    MUNDO   "))  # 'hola mundo'
````

## 6️⃣ Alcance básico de variables (scope)
- **Local:** definida dentro de una función.
- **Global:** definida fuera de funciones.

> Las funciones pueden **leer** variables globales; para **modificarlas**, usa `global`.


````python
# Ejemplos de alcance
x = 5  # global

def leer_global():
    print("leer_global -> x =", x)

def modificar_global():
    global x
    x = x + 1

leer_global()
modificar_global()
leer_global()
````

## 7️⃣ Buenas prácticas con funciones
- **Responsabilidad única:** cada función debe hacer una cosa bien.
- **Nombres descriptivos:** verbos para acciones (`calcular_total`).
- **Docstrings:** describe propósito, parámetros y retorno.
- **No abusar de globales:** preferir parámetros y retornos.
- **Probar funciones pequeñas:** con casos simples y bordes.
